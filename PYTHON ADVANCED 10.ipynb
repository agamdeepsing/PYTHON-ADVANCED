# Ques 3

__getattr__(self, name) is only called when an attribute is not found in the usual places (i.e., in the instance's dictionary or in its class hierarchy). It is a fallback method that is called when the attribute access fails with an AttributeError. This method takes a single argument, name, which is the name of the attribute being accessed, and should return the value of the attribute.

On the other hand, __getattribute__(self, name) is called every time an attribute is accessed, regardless of whether the attribute exists or not. This method takes two arguments, self and name, and should return the value of the attribute. If you don't implement this method, the default behavior is to simply look up the attribute in the instance's dictionary.



properties are defined on a per-instance basis, while descriptors are defined on the class. When you define a property, it is essentially a shorthand for defining a descriptor that only implements the __get__() method (and optionally the __set__() and __delete__() methods). This means that properties can have different values for each instance of a class, while descriptors are shared across all instances.
