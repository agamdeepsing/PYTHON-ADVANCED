In Python, both __getattr__ and __getattribute__ are special methods that allow to customize attribute access for objects.__getattribute__ is used to find an attribute of a class.__getattr__ is implemented latter if AttributeError is generated by __getattribute__, but for this __getattribute__ and __getattr__ both has to be defined in same class.

#In general, be careful when using __getattribute__, as it can potentially cause infinite recursion if  not careful.only use it when you really need to intercept every attribute access.

# Ques 2

In Python, both properties and descriptors are mechanisms for customizing attribute access on objects
With Properties we can bind getter, setter and delete functions together with an attribute name, using the built-in property function or @property decorator.

With Descriptor we can bind getter, setter and delete functions into a seperate class.
